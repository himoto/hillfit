```python
from numpy import array
import pandas
from hillfit import HillFit

# import and prepare the modeling data
data = pandas.read_csv(open(r'processed_data.csv'))
x = array([x for x in data['time (hr)'] if x != 0])
y = array([y for y in data['oxidation_proportion'] if y != 0])

# execute the Hillfit module
%matplotlib inline

model = HillFit(x, y)
x_label = 'hours'
y_label = 'reduction'
title = 'Fitted Hill equation for porphyrin PDI'
x_fit, y_fit, params, equation = model.fitting(x_label, y_label, title)
model.export()
print(equation)
```
